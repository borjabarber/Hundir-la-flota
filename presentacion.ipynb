```python

import numpy as np
import random 
```

```python


def crear_tablero(tamano=10):
    return np.full((tamano, tamano), "_")  #np.full crea la matriz 

print(crear_tablero()) 
```

```python
def colocar_barco(barco, tablero): 
    for casilla in barco:  #casilla es una coordenada en el tablero.
        fila, columna = casilla #descompone esa coordenada en dos partes y los mete en la variable fila y una columna.
        tablero[fila, columna] = 'O'  #coloca una "O" en la posición indicada por fila y columna dentro del tablero 

#modo de uso 
tablero = crear_tablero()
colocar_barco([(0, 1), (1, 1)], tablero)
colocar_barco([(1, 3), (1, 4), (1, 5), (1, 6)], tablero)
print(tablero) 
```

```python
def disparar(casilla, tablero):
    fila, columna = casilla
    if tablero[fila, columna] == "O": # comprueba si en el tablero, en la posición indicada por fila y columna, hay un barco ("O").
        tablero[fila, columna] = "X" # si lo hay marca la casilla con una "X" y retorna tocado 
        return "Tocado" 
    else:
        tablero[fila, columna] = "A" # si no marca con una "A" y retorna agua 
        return "Agua"
    
    
#modo de uso 
resultado = disparar((0, 4), tablero)
print(tablero)
print(resultado)
```

```python
#modo random 

#crear barcos random 

def crear_barco(eslora, tamaño_tablero=10): #crea un barco de una longitud específica (eslora) en el tamaño del tablero 
    orientacion = random.choice(['horizontal', 'vertical']) #random.choice para que elija entre vertical y horizontal 
    if orientacion == 'horizontal':
        fila = random.randint(0, tamaño_tablero - 1) #random.radint Elige una fila aleatoria en el tablero 
        columna_inicial = random.randint(0, tamaño_tablero - eslora) #Elige una columna inicial de manera que el barco quepa dentro del tablero.
        barco = [(fila, columna_inicial + i) for i in range(eslora)] #Genera una lista de coordenadas (fila, columna)
                                                                     #si el barco tiene 3 casillas y empieza en la columna 2, 
                                                                     #las coordenadas serían [(fila, 2), (fila, 3), (fila, 4)].
    else:
        fila_inicial = random.randint(0, tamaño_tablero - eslora) #lo mismo pero para vertical
        columna = random.randint(0, tamaño_tablero - 1) 
        barco = [(fila_inicial + i, columna) for i in range(eslora)] #la i es el tamaño del barco usando eslora como guia 
    return barco

#modo de uso 
barco = crear_barco(4)
print(barco)
```

```python

# colocar barcos random


#creo una funcion que me devuelva un boleano que usare en un bucle while not despues 
def puede_colocar_barco(barco, tablero):#Esta función verifica si el barco puede ser colocado en las 
                                         #posiciones dadas del tablero (es decir, si las casillas están vacías).
    for casilla in barco:
        fila, columna = casilla
        if tablero[fila, columna] != "_":
            return False
    return True


def colocar_barcos(tablero):
    barcos = [] # creamos una lista vacia de barcos para meter el conjunto de barcos que bamos a crear a continuacion 
    barcos.extend([crear_barco(2) for _ in range(3)]) #uso la funcion crear barco para el tamaño del barco y rango para el numero de barcos  
    barcos.extend([crear_barco(3) for _ in range(2)]) #uso extend para el conjunto de barcos (cuando son mas de uno)
    barcos.append(crear_barco(4)) #uso append cuando solo es un barco 
    
    for barco in barcos: #para que no se solapen los barcos recorro cada elemento en la lista barcos y hago la comprobacion con while not 
        while not puede_colocar_barco(barco, tablero):#con while not me aseguro de que el barco se coloque sin superponerse a otros barcos
            barco = crear_barco(len(barco)) #Mientras no se puede colocar el barco en las posiciones actuales, se crea un nuevo barco con la misma longitud.
        colocar_barco(barco, tablero) #Una vez que se encuentra una posición válida donde el barco no se superpone con otros, se coloca el barco en el tablero.



#modo de uso 
tablero = crear_tablero()
colocar_barcos(tablero)
for fila in tablero:
    print(' '.join(fila)) 
```

```python
tamano_tablero = 10

tablero_jugador = crear_tablero(tamano_tablero)
colocar_barcos(tablero_jugador)

tablero_skynet = crear_tablero(tamano_tablero)
colocar_barcos(tablero_skynet)

barcos_restantes_jugador = True
barcos_restantes_skynet = True

turno_actual = 1
turnos_totales = 10

while barcos_restantes_jugador and barcos_restantes_skynet and turno_actual <= turnos_totales: 
    print("Turno" , turno_actual, "/" ,turnos_totales,)

    print("Tu tablero")
    print(tablero_jugador)
    
    # turno del jugador 
    
    #para comprobar si el disparo es valido sin interumpir ni dar errores 
    disparo_valido = False # creo la variable disparo_valido = False para controlar la validez de la entrada del jugador. 
    while not disparo_valido: # El bucle while continuará ejecutándose hasta que disparo_valido sea True.
        casilla = input("Introduce las coordenadas de tu disparo (fila y columna): ") #Solicita al jugador que introduzca las coordenadas de su disparo
        
        #try y except me Permiten manejar errores de manera elegante sin que el programa se detenga abruptamente.
        try: 
            fila, columna = map(int, casilla.split(",")) #Intenta dividir la entrada del usuario en dos partes (fila y columna) y convertirlas en enteros.
            if 0 <= fila < tamano_tablero and 0 <= columna < tamano_tablero: # Verifica si las coordenadas están dentro de los límites del tablero.
                disparo_valido = True #si todo el correcto el disparo es valido 
            else: #si no 
                print("Señor, Cordenadas fuera del rango del tablero. Intenta nuevamente.") #tira el mensaje de error 
        except ValueError: #Maneja cualquier error que ocurra durante la entrada del usuario a enteros.
            print("Señor, asegúrese de introducir números separados por una coma. Intenta nuevamente.")
    
    #para saber si el dispare a hacertado 
    resultado = disparar((fila, columna), tablero_skynet) #La función disparar determina si el disparo fue exitoso ("Tocado") o no ("Agua"). y lo guarda
    print(resultado) #para que se vea el resultado 

    # Verificar si quedan barcos en el tablero de la máquina
    barcos_restantes_skynet = np.any(tablero_skynet == "O") #la función np.any es el como una funcion bolleana de numpy 
    if not barcos_restantes_skynet:
        print("Has salvado el futuro")
        break
    
    #turno de la maquina 
    
    # Imprimir los disparos de la máquina
    fila_skynet = random.randint(0, tamano_tablero - 1)
    columna_skynet = random.randint(0, tamano_tablero - 1)
    resultado_skynet = disparar((fila_skynet, columna_skynet), tablero_jugador)
    print(f"el T-1000 ha ordenado un disparo en ({fila_skynet}, {columna_skynet}): {resultado_skynet}") # uso la f para decirle a python que dentro de ese string 
                                                                                                        # puede haber variables dentro de las llaves

    # Verificar si quedan barcos en el tablero del jugador
    barcos_restantes_jugador = np.any(tablero_jugador == "O")
    if not barcos_restantes_jugador:
        print("Jhon Connor ha muerto")
        break

    turno_actual += 1 

if turno_actual > turnos_totales:
    print("Ha explotado una bomba nucelar. El juego ha terminado.")
```

¿Por qué usar while not y try/except?
while not: Asegura que el programa siga pidiendo una entrada válida hasta que se cumplan las condiciones adecuadas. Esto es útil para evitar que el juego procese coordenadas incorrectas, lo que podría llevar a errores o comportamientos inesperados.

try y except: Permiten manejar errores de manera elegante sin que el programa se detenga abruptamente. Cuando un jugador introduce datos incorrectos, el bloque except captura el error y permite que el programa informe al jugador y le pida que intente nuevamente.
